# PyTorch Model Creation, Saving, and Loading Example

This notebook demonstrates the basic workflow for creating a simple deep learning model in PyTorch, saving its trained weights, and then loading those weights to make predictions, simulating a transfer to a new environment.

## Part 1: Create and Train a Model, then Save Weights

```python
# Import necessary PyTorch libraries
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
```

```python
# 1. Define a simple PyTorch model (e.g., a simple neural network)
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(10, 5) # Input features: 10, Output features: 5
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(5, 1)  # Output features: 1 (for a regression-like task)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

print("Model defined.")
```

```python
# 2. Create an instance of the model
model = SimpleNN()

# Create a dummy dataset for demonstration
# Let's assume input features are 10 and output is 1
X_train = torch.randn(100, 10) # 100 samples, 10 features
y_train = torch.randn(100, 1)  # 100 corresponding labels

# Define a loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# A brief training loop (for demonstration of weight changes)
num_epochs = 10
print(f"Starting a brief training for {num_epochs} epochs...")
for epoch in range(num_epochs):
    # Forward pass
    outputs = model(X_train)
    loss = criterion(outputs, y_train)

    # Backward and optimize
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch+1) % 2 == 0:
        print (f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

print("Brief training complete.")
```

```python
# 3. Save the model's state dictionary (weights and biases) to a .pth file
# This saves only the learnable parameters, not the model architecture.
model_weights_path = 'simple_nn_weights.pth'
torch.save(model.state_dict(), model_weights_path)

print(f"Model weights saved to {model_weights_path}")
```

## Part 2: Recreate Model, Load Weights, and Predict (Simulating a New Platform)

Imagine this section is run on a completely different system or at a later time. We only have the saved `.pth` file and the knowledge of the model's architecture.

```python
# Import necessary PyTorch libraries (again, to simulate a fresh environment)
import torch
import torch.nn as nn

# 1. Re-define the *exact same* model architecture
# It's crucial that this architecture matches the one used during training.
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(10, 5)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(5, 1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

# Create a new instance of the model
loaded_model = SimpleNN()
print("New model instance created.")
```

```python
# 2. Load the saved state dictionary into the new model instance
model_weights_path = 'simple_nn_weights.pth'
loaded_model.load_state_dict(torch.load(model_weights_path))

# Set the model to evaluation mode
loaded_model.eval() # This is important for layers like BatchNorm and Dropout

print(f"Weights loaded from {model_weights_path}. Model is now in evaluation mode.")
```

```python
# 3. Perform a prediction task with the loaded model
# Create some new, unseen data
X_new = torch.randn(5, 10) # 5 new samples, 10 features

print("New input data:")
print(X_new)

# Make predictions (ensure no gradient calculation for inference)
with torch.no_grad():
    predictions = loaded_model(X_new)

print("\nPredictions from the loaded model:")
print(predictions)
```
